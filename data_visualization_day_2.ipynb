```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### import the titanic dataset

```python
df=pd.read_csv("titanic.csv")
df
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
### code here
#plt.figure(figsize=(6,7))
plt.bar(df["Fare"],height=7,data=df)
plt.title("Fare of Different Classes")
plt.show
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
### code here
plt.barh(y="Fare",height=10,width=4,data=df)
plt.title("Fare of Different Classes")
plt.ylabel("Fare")
plt.xlabel("Class")
plt.show
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
### code here
colors = ['red', 'blue', 'green', 'orange']
plt.bar(x="Fare",height=10,width=4,color=colors,data=df)
plt.title("Fare of Different Classes")
plt.xlabel("Fare")
plt.ylabel("Class")
plt.show
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
a=df["Fare"].value_counts()
```

```python
df["Fare"].value_counts().plot.pie()
```

```python
### code here
plt.figure(figsize=(20,10))
plt.pie(df["Fare"].value_counts(),labels=a.index,autopct="%.1ff%%")
plt.show
```

###### 6. Write a Python program to display a factor plot of ( pclass vs embarked). hist (kind = 'count')

```python
#### code here
sns.factorplot(x="Pclass",col="Embarked",kind="count",data=df)
plt.show
```
