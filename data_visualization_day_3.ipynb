```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##### import tips dataset.

```python
df=pd.read_csv("tips.csv")
df
```

##### 1. Write a Python program to display a bar chart of the total bill of Gender and write the intuition about the plot.

####  
`The main intuition behind a bar chart is to visually display the magnitude or frequency of each category by the length or height of the corresponding bar`

```python
### code here
plt.figure(figsize=(7,6))
plt.bar(df["sex"],df["total_bill"],data=df)
plt.title("Total Bill by Gender")
plt.xlabel("Sex")
plt.ylabel("Total Bill")
plt.xticks(rotation=45)
plt.show
```

###### 2. Write a Python program to display a box plot of the total bill of day and hue = "smoker" and write the intuition about the plot.

#### 
`The main intuition behind a box plot is to represent the key characteristics of the data distribution in a compact and easy-to-understand format. It consists of a rectangle (the box) and two lines (the whiskers) that extend from the box.`

```python
#### code here
plt.figure(figsize=(7,6))
sns.boxplot(x="day",y="total_bill",hue="smoker",data=df)
plt.title("Total Bill by Day with Smoker Hue")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.xticks(rotation=45)
plt.show
```

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

####  
`The main intuition behind a violin plot is to visualize the distribution of data and compare it across different categories or groups`

```python
### code here
plt.figure(figsize=(7,6))
sns.violinplot(x="day",y="total_bill",hue="sex",data=df)
plt.title("Total Bill by Day with sex Hue")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.xticks(rotation=45)
plt.show
```

##### 4. Write a Python program to display a strip plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

#####   
`The main intuition behind a strip plot is to visualize the individual data points and their distribution along a continuous scale. Each dot represents a single data point and is positioned according to its value on the axis`

```python
### code here 
plt.figure(figsize=(7,6))
sns.stripplot(x="day",y="total_bill",hue="sex",data=df)
plt.title("Total Bill by Day with sex Hue")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.xticks(rotation=45)
plt.show
```

##### 5. Write a Python program to display heat map of dataframe and write the intuition about the plot.

#####  
`The main intuition behind a heatmap plot is to visually depict the magnitude or intensity of values across two dimensions. It allows for easy identification of areas of high and low values, as well as patterns or clusters within the data.`

```python
df.corr()
```

```python
### code here
plt.figure(figsize=(7,6))
sns.heatmap(df.corr(),data=df,annot=True)
plt.title("Heat Map of DataFrame")
plt.xlabel("X-axis DataFrame")
plt.ylabel("Y-axis DataFrame")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show
```

##### 6. Write a Python program to display scatter plot of total bill of tip and write the intuition about the plot.

#### `The main intuition behind a scatter plot is to visually examine the relationship or correlation between two variables`

```python
### code here
plt.figure(figsize=(7,6))
sns.scatterplot(df["tip"],df["total_bill"],data=df)
plt.title("Total Bill by tip")
plt.xlabel("Tip")
plt.ylabel("Total Bill")
plt.show
```

##### 7. Write a Python program to display density plot of all numerical columns from the dataframe and write the intution about plot.

###### 
**`A density plot, also known as a kernel density plot, is a data visualization technique that provides insights into the distribution and shapeof a continuous variable`**

```python
sns.kdeplot(df['total_bill'])
plt.show
```

```python
sns.kdeplot(df['tip'])
plt.show
```

```python
sns.kdeplot(df['size'])
plt.show
```

```python
df[['total_bill','tip']].plot.kde()
plt.title("Density plot of all numerical column ")
plt.xlabel("Total Bill")
plt.ylabel("Density of Tips")
plt.show
```

##### 8. Write a Python program to calculate average total bills by days and display top 4 days by plot.

```python
### code here
df.groupby("day")["total_bill"].mean().head(4).plot.bar()
plt.title("Average total Bill by Days")
plt.xlabel("Days of the Week")
plt.ylabel("average of Total Bill")
plt.xticks(rotation=45)
plt.show
```
