## Line Plot

```python
plt.figure(figsize=(9,5))
sns.lineplot(x="carat",y="depth",data=df)
plt.title("Carat vs Depth")
plt.xlabel("Carat")
plt.ylabel("Depth of the carat")
plt.show
```

```python
plt.figure(figsize=(9,5))
sns.lineplot(x="table",y="depth",data=df)
plt.title("table vs Depth")
plt.xlabel("Carat")
plt.ylabel("Depth of the table")
plt.show
```

```python
plt.figure(figsize=(9,5))
sns.lineplot(x="x",y="y",data=df)
plt.title("x vs y")
plt.xlabel("Type x")
plt.ylabel("Type Y ")
plt.show
```

## Heat Map

```python
plt.figure(figsize=(9,5))
colors=sns.color_palette("rocket")
sns.heatmap(df[["carat","depth"]].corr(),cmap=colors,annot=False)
plt.title("Carat vs Depth")
plt.xlabel("Carat")
plt.ylabel("Depth ")
plt.show
```

```python
plt.figure(figsize=(9,5))
colors=sns.color_palette("mako")
sns.heatmap(df[["depth","table"]].corr(),cmap=colors,annot=False)
plt.title("depth vs table")
plt.xlabel("depth")
plt.ylabel("table ")
plt.show
```
