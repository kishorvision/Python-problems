###### 1. load the iris.csv dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Iris.csv")
df
```

```python
df["Species"].value_counts()
```

#### 1. create a bar plot for count the frequency of the three species.

```python
### code here
plt.figure(figsize=(7,5.5))
sns.countplot(x="Species",data=df)
plt.title("Iris Species")
plt.ylabel("Frequency of the Species")
plt.xlabel("Type of Species")
plt.xticks(rotation=(45))
plt.show
```

##### 2. Create a pie chart to show the percentage of species.

```python
### code here
plt.figure(figsize=(7,5.5))
plt.pie(df["Species"].value_counts(),labels=df["Species"].value_counts().index,autopct="%.1f%%")
plt.show
```

##### 3. Create the joint to all numberical variable for analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
### code here
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",kind="scatter",data=df)
plt.show
```

##### 4. Create a scatter plot and take hue = "Species"

```python
### code here
plt.figure(figsize=(9,5))
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",hue="Species",data=df)
plt.title("SepalLengthCm vs SepalWidthCm")
plt.show
```

###### 5. Create a Boxplot or Whisker plot as univeriate and bivariate analysis.

### Univariate Box Plot

```python
### code here
plt.figure(figsize=(5,2))
sns.boxplot(x="PetalLengthCm",data=df)
plt.title("Univariate Box Plot for Iris Dataset")
plt.show
```

### Bivariate Box Plot

```python
plt.figure(figsize=(5,4))
sns.boxplot(x="Species",y="PetalLengthCm",data=df)
plt.title("Bivariate Box plot for Iris Dataset")
plt.show
```

###### 6. Create a Strip plot 

```python
### code here
sns.stripplot(x="PetalLengthCm",y="Species",hue="Species",marker="v",s=5,data=df)
plt.title("Strip Plot")
plt.yticks(rotation=45)
plt.show
```

###### 7. Create a Violin Plot.

```python
### code here
plt.figure(figsize=(8,5))
sns.violinplot(x="Species",y="PetalWidthCm",data=df)
plt.title("VIOLINE PLOT")
plt.xlabel("Species")
plt.xticks(rotation=45)
plt.show
```

##### 8. Create a pair chart.

```python
### code here
sns.pairplot(data=df,vars=["SepalLengthCm","SepalWidthCm","PetalLengthCm"])
plt.show
```

```python
sns.pairplot(data=df,vars=["SepalLengthCm","SepalWidthCm","PetalLengthCm"],kind="kde")
plt.show
```

##### 9. Create a Heat Map

```python
### code here
plt.figure(figsize=(9,6))
colors=sns.color_palette("Spectral")
sns.heatmap(df.corr(),cmap=colors,annot=True)
plt.title("Heat Map")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show
```
